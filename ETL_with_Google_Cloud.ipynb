## Data Cleaning

```
df.sample(5) #Orginal Data 
```

```
df['Day'] = pd.to_datetime(df['Day'])
```

```
df['Impressions'] = df['Impressions'].apply(lambda x: int(x.replace(',','')))
```

```
df['Cost'] = df['Cost'].apply(lambda x: float(x.replace(',','')))
```

```
df['Conv. value'] = df['Conv. value'].apply(lambda x: float(x.replace(',','')))
```
